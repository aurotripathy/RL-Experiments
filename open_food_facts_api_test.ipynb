```python
# https://openfoodfacts.github.io/openfoodfacts-python/usage/

from openfoodfacts import API, APIVersion, Country, Environment, Flavor

api = API(
    user_agent="MyAwesomeApp/1.0",
    username=None,
    password=None,
    country=Country.us,
    flavor=Flavor.off,
    version=APIVersion.v2,
    environment=Environment.org,
    timeout=40, # looks to be the minimum timeout
)
```

```python
code = "3017620422003"
try:
    result = api.product.get(code, fields=["code", "product_name"])
    print(result)
    # {'code': '3017620422003', 'product_name': 'Nutella'}
except Exception as e:
    print(f"Error fetching product: {e}")
```

```python
import pprint
try:
    result = api.product.text_search("KitKat")
    pprint.pprint(result)
    # {"count": 3006628, "page": 1, "page_count": 20, "page_size": 20, "products": [{...}], "skip": 0}
except Exception as e:
    print(f"Error searching products: {e}")
```
